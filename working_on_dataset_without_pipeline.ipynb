WORKING ON TITANIC DATASET WITHOUT PIPELINES

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer #for missing data
from sklearn.preprocessing import OneHotEncoder #nominal data
from sklearn.preprocessing import MinMaxScaler #for scaling
from sklearn.tree import DecisionTreeClassifier #for prediction
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
#step 1 == train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df.drop(columns=['Survived']),
                                                 df['Survived'],
                                                 test_size=0.2,
                                                 random_state=42)
```

```python
X_train.head(2)
```

```python
Y_train.head()
```

```python
df.isnull().sum()
```

```python
#Applying imputation

si_age = SimpleImputer() #default is using mean
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
X_train_age
```

```python
X_train_age
```

```python
# One hot encoding on sex and embarked

ohe_sex = OneHotEncoder(sparse=False,handle_unknown='ignore') #handle_unknown means ignore completely new category
ohe_embarked = OneHotEncoder(sparse=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train[['Embarked']])

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test[['Embarked']])
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
X_train.head(5)
```

```python
X_train_rem = X_train.drop(columns=['Age','Sex','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Age','Sex','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_sex,X_train_embarked),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_sex,X_test_embarked),axis=1)
```

```python
X_train_transformed
```

```python
X_train_transformed.shape
```

```python
X_test_transformed
```

```python
X_test_transformed.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,Y_train)
```

```python
Y_pred = clf.predict(X_test_transformed)
Y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```
